```python
from bs4 import BeautifulSoup

import urllib.request

response = urllib.request.urlopen('http://php.net/')

html = response.read()

soup = BeautifulSoup(html,"html5lib")

text = soup.get_text(strip=True)

print (text)
```

```python
nltk.download('punkt')
```

```python
word_tokens = nltk.word_tokenize(text)
print(word_tokens)
freq = nltk.FreqDist(word_tokens)
```

```python
for key,val in freq.items():

    print (str(key) + ':' + str(val))
```

```python
freq.plot(20, cumulative=False)
```

```python
from nltk.tokenize import sent_tokenize

mytext = "Buenas tardes, señorita ¿en qué puedo servirles? Buenas tardes ¿puede decirme que tiene la hamburguesa regular? Por supuesto, tiene 200gr de carne, queso cheddar, tocino, rodajas de tomate y salsas.-Cliente: Excelente. Quiero una, acompañada de una gaseosa.-Camarero: Muy bien. ¿Querrá algún postre con la comida? Aún no lo sé. Te diré cuando traigas la hamburguesa. De acuerdo. En un momento le traeré su pedido. Muchas gracias."


sent_tokenize(mytext,language='spanish')
```

```python
from nltk.corpus import wordnet

wordnet.langs()
synonyms = []


for syn in wordnet.synsets("malo", lang='spa'):
    for l in syn.lemmas(lang='spa'):
        synonyms.append(l.name())
print(set(synonyms))


antonyms = []

for syn in wordnet.synsets("padre", lang='spa'):

    for l in syn.lemmas():

        if l.antonyms():

            antonyms.append(l.antonyms()[0].name())

print(set(antonyms))
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

print(stemmer.stem('played'))
```

```python
from nltk.stem import SnowballStemmer

print(SnowballStemmer.languages)

stemmer = SnowballStemmer(language='english')

print(stemmer.stem("running"))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('increases'))
word= 'running'



#verbo, sustantivo, adjetivo o adverbio:

print(lemmatizer.lemmatize(word, pos="v"))

print(lemmatizer.lemmatize(word, pos="n"))

print(lemmatizer.lemmatize(word, pos="a"))

print(lemmatizer.lemmatize(word, pos="r"))
```
